```python
d6 = []
#ef.loc[38]['Country Name']


for i in range(len(ef.index)):
    
    col = ef.index[i]
    country_name = ef.loc[ef.index[i]][ef.columns[0]]
    for j in range(2 , len(ef.columns)):
        seged = {}
        row = ef.columns[j]
        
        seged['country_name'] = country_name
        seged['year'] = ef.columns[j]
        seged['val'] = ef.loc[col][row]
        
        #print(seged)
        
        d6.append(seged)
        

# Export to json
import json

with open('bigdata/projekt.json', 'w+') as f:
    json.dump(d6, f);
```
